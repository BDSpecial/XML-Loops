```python
# convert the 'PressureInjuryDateofSurvey' and 'RestraintDateofSurvey' columns to datetime format
HAPI_Restraint_Summary.PressureInjuryDateofSurvey = pd.to_datetime(HAPI_Restraint_Summary.PressureInjuryDateofSurvey)
HAPI_Restraint_Summary.RestraintDateofSurvey = pd.to_datetime(HAPI_Restraint_Summary.RestraintDateofSurvey)
# Change the date formating
HAPI_Restraint_Summary.PressureInjuryDateofSurvey = HAPI_Restraint_Summary.PressureInjuryDateofSurvey.apply(lambda x: x.strftime('%m/%d/%Y'))
HAPI_Restraint_Summary.RestraintDateofSurvey = HAPI_Restraint_Summary.RestraintDateofSurvey.apply(lambda x: x.strftime('%m/%d/%Y'))


# Change the patient age from float to integer but first make the null age a 0
HAPI_Details.PatientAge = HAPI_Details.PatientAge.fillna(0.0).astype(int)
Restraint_Details.PatientAge = Restraint_Details.PatientAge.fillna(0.0).astype(int)
# Change the patient age from float to integer but first make the null age a 0
HAPI_Details.xRestraintUlcerAgeMonthsFID = HAPI_Details.xRestraintUlcerAgeMonthsFID.fillna(0.0).astype(int)
Restraint_Details.xRestraintUlcerAgeMonthsFID = Restraint_Details.xRestraintUlcerAgeMonthsFID.fillna(0.0).astype(int)
# Change the patient age from float to integer but first make the null age a 0
HAPI_Details.xRestraintUlcerAgeDaysFID = HAPI_Details.xRestraintUlcerAgeDaysFID.fillna(0.0).astype(int)
Restraint_Details.xRestraintUlcerAgeDaysFID = Restraint_Details.xRestraintUlcerAgeDaysFID.fillna(0.0).astype(int)
# Change the patient age from float to integer but first make the null age a 0
HAPI_Details.xPressureUlcerGestationalAgeFID = HAPI_Details.xPressureUlcerGestationalAgeFID.fillna(0.0).astype(int)


# Change the Branden score from float to integer but first make the null score a 0
HAPI_Details.xUlcerRiskAssessmentScoreOnAdmission = HAPI_Details.xUlcerRiskAssessmentScoreOnAdmission.fillna(0.0).astype(int)
HAPI_Details.xUlcerRiskAssessmentScoreLast = HAPI_Details.xUlcerRiskAssessmentScoreLast.fillna(0.0).astype(int)
# Change the TotalUlcerCount from float to integer but first make the null score a 0
HAPI_Details.TotalUlcerCount = HAPI_Details.TotalUlcerCount.fillna(0.0).astype(int)

# Change the Branden score from float to integer but first make the null score a 0
Restraint_Details.TypeLimb = Restraint_Details.TypeLimb.fillna(0.0).astype(int)
Restraint_Details.TypeVest = Restraint_Details.TypeVest.fillna(0.0).astype(int)
Restraint_Details.TypeOther = Restraint_Details.TypeOther.fillna(0.0).astype(int)
Restraint_Details.JustifyFall = Restraint_Details.JustifyFall.fillna(0.0).astype(int)
Restraint_Details.JustifyRemoveEquipment = Restraint_Details.JustifyRemoveEquipment.fillna(0.0).astype(int)
Restraint_Details.JustifyHarmSelf = Restraint_Details.JustifyHarmSelf.fillna(0.0).astype(int)
Restraint_Details.JustifyHarmOthers = Restraint_Details.JustifyHarmOthers.fillna(0.0).astype(int)
Restraint_Details.JustifyOther = Restraint_Details.JustifyOther.fillna(0.0).astype(int)
Restraint_Details.JustifyUnknown = Restraint_Details.JustifyUnknown.fillna(0.0).astype(int)
```

```python
# The variables needed for this function are the: Hospital abbreviation, Hospital NDNQI template, Hospital NDNQI
# code, Calendar year being submitted, and Calendar quarter being submitted
def NDNQI_HAPI_Restraint_Summary_XML(Selected_Hosp,Hosp_XML_Template,Hosp_NDNQI_Code,Calendar_Year,Calendar_Quarter):
    # This section is to update the total Census and Survey counts
    
    # This Hospital's HAPI_Restraint Summary Data
    Hosp_data = HAPI_Restraint_Summary[HAPI_Restraint_Summary.Hospital == Selected_Hosp]
    # A List of this Hospital's Unit Names
    Hosp_Units = list(set(Hosp_data.UnitName))

    #This is the current XML template for this hospital and their NDNQI units for this metric
    tree = xml.dom.minidom.parse(Hosp_XML_Template)
    # This is each unit node of the xml file
    unit_nodes = tree.getElementsByTagName("UnitYrQtr")

    # Make sure the correct year and quarter are listed
    H = tree.getElementsByTagName("HospitalCode")[0].childNodes[0].nodeValue = Hosp_NDNQI_Code
    Y = tree.getElementsByTagName("Year")[0].childNodes[0].nodeValue = Calendar_Year
    Q = tree.getElementsByTagName("Quarter")[0].childNodes[0].nodeValue = Calendar_Quarter


    # This loops through each separate UnitName from the Template XML file and replaces it WITH the correct NDNQI
    # name for the units.  
    for idx, unit in enumerate(unit_nodes):
        unit.getElementsByTagName('UnitName')[0].childNodes[0].nodeValue = Hosp_Units[idx]

        # Now grab each of the Summary Data for that unit from the Hospital data
        new_data = Hosp_data[Hosp_data.UnitName == Hosp_Units[idx]]

        # Take each xml Month and replace it with each listed month from the new_data for that unit
        month_nodes = unit.getElementsByTagName("UnitMonth")
        for idx2, month in enumerate(month_nodes):
            month.getElementsByTagName('Month')[0].childNodes[0].nodeValue = new_data.Month.iloc[idx2]

            # Now replace each xml metric value with data that has the same name from our data source
            for column in new_data.columns[3:]:

                new_value = new_data[column].iloc[idx2]
                month.getElementsByTagName(column)[0].childNodes[0].nodeValue = new_value



    # This is a temporary xml file that I will now use to add the individual separation data.
    tree.writexml(open('Temp_{}.xml'.format(Selected_Hosp), 'w'))
```

```python
# The variables needed for this function are the: Hospital abbreviation, Hospital Temp XML file,
# Calendar year being submitted, and Calendar quarter being submitted
def NDNQI_Indiv_HAPI_Restraint_XML(Selected_Hosp,Temp_XML,Calendar_Year,Calendar_Quarter):
    # This section is to insert the individual HAPI and Restraint patients
    
    # This Hospital's last Quarter's individual HAPI details
    Hosp_individual_restraints = Restraint_Details[Restraint_Details.Hospital == Selected_Hosp]
    # This Hospital's last Quarter's individual HAPI details
    Hosp_individual_HAPI = HAPI_Details[HAPI_Details.Hospital == Selected_Hosp]


    individual_HAPI = ["UlcerIdentifier", "PatientAge", "xRestraintUlcerAgeMonthsFID","xRestraintUlcerAgeDaysFID",
    "xPressureUlcerGestationalAgeFID", "xGenderFID", "xRecencyHoursFID", "xPressureUlcerSkinAssessmentOnAdmissionFID",
    "xPressureUlcerRiskAssessmentOnAdmissionFID", "xUlcerRiskAssmntFIDOnAdmissionScale", "xUlcerRiskAssessmentScoreOnAdmission",
    "xUlcerAtRiskLastAssessmentFID", "xUlcerRiskAssmntFIDLastScale", "xUlcerRiskAssessmentScoreLast",
    "TotalUlcerCount", "HospAcquiredCount", "StageICount", "StageIICount", "StageIIICount", "StageIVCount",
    "DtiCount", "StageUnstageableEscharSlough", "MucosalMembCount", "NonvisibleCount", "MedicalDeviceCount",
    "xPressureUlcerPreventionLast24HoursFID", "xUlcerInterventionSkinAssessmentDocumentedFID",
    "xUlcerInterventionPressureRedistributionSurfaceFID","xUlcerInterventionRoutineRepositioningFID",
    "xUlcerInterventionNutritionalsupportFID", "xUlcerInterventionMoistureManagementFID", "UnitAcquiredCount",
    "StageICountUnit", "StageIICountUnit", "StageIIICountUnit", "StageIVCountUnit", "DtiCountUnit",
    "StageUnstageableEscharSloughUnit", "MucosalMembCountUnit", "NonvisibleCountUnit", "MedicalDeviceCountUnit"]

    individual_restraint = ["RestraintIdentifier", "PatientAge", "xRestraintUlcerAgeMonthsFID", 
    "xRestraintUlcerAgeDaysFID","xGenderFID", "xRestraintInUseFID", "TypeLimb", "TypeVest", "TypeOther",
    "xRestraintCategoryFID", "JustifyFall", "JustifyRemoveEquipment", "JustifyHarmSelf", "JustifyHarmOthers", 
    "JustifyOther", "JustifyUnknown"]


    # This is the temporary xml file that has already had the Unit Census and date section filled
    tree = xml.dom.minidom.parse(Temp_XML)
    unit_nodes = tree.getElementsByTagName("UnitYrQtr")

    # This loops through each listed unit in our xml file
    for idx, unit in enumerate(unit_nodes):
        each_unit = unit.getElementsByTagName('UnitName')[0].childNodes[0].nodeValue

        # Then we loop through each listed month
        month_nodes = unit.getElementsByTagName("UnitMonth")
        for idx2, month in enumerate(month_nodes):

            # Then I grab each patient from Hosp_individual_HAPI data for this unit
            HAPI_patients = Hosp_individual_HAPI[Hosp_individual_HAPI.UnitName == each_unit]

            Restraint_patients = Hosp_individual_restraints[Hosp_individual_restraints.UnitName == each_unit]

            # Now I can take each listed HAPI patient
            for i in range(len(HAPI_patients)):
                # and create a "UnitUlcer" node for each individual patient
                HAPI_patient_node = tree.createElement("UnitUlcer")
                for data in individual_HAPI:
                    # I create an element for each individual's metrics
                    data_name = tree.createElement(data)

                    if data in ["xGenderFID", "xRecencyHoursFID", "xPressureUlcerSkinAssessmentOnAdmissionFID",
    "xPressureUlcerRiskAssessmentOnAdmissionFID", "xUlcerRiskAssmntFIDOnAdmissionScale", "xUlcerRiskAssessmentScoreOnAdmission",
    "xUlcerAtRiskLastAssessmentFID", "xUlcerRiskAssmntFIDLastScale","xPressureUlcerPreventionLast24HoursFID", "xUlcerInterventionSkinAssessmentDocumentedFID",
    "xUlcerInterventionPressureRedistributionSurfaceFID","xUlcerInterventionRoutineRepositioningFID",
    "xUlcerInterventionNutritionalsupportFID", "xUlcerInterventionMoistureManagementFID"] and pd.isna(HAPI_patients[data].iloc[i]):
                        data_name.appendChild(tree.createTextNode(''))
                    elif data in ["PatientAge", "xRestraintUlcerAgeMonthsFID",
                                  "xRestraintUlcerAgeDaysFID","xPressureUlcerGestationalAgeFID",] and HAPI_patients[data].iloc[i] == 0:
                        data_name.setAttribute('xsi:nil', 'true')
                    elif data in ["xUlcerRiskAssessmentScoreOnAdmission", "xUlcerRiskAssessmentScoreLast"] and HAPI_patients[data].iloc[i] == 0:
                        data_name.appendChild(tree.createTextNode(''))
                    elif pd.isna(HAPI_patients[data].iloc[i]):
                        data_name.setAttribute('xsi:nil', 'true')
                    else:
                        # and I add each one to that individual's "UnitUlcer" node
                        data_name.appendChild(tree.createTextNode(str(HAPI_patients[data].iloc[i])))

                    HAPI_patient_node.appendChild(data_name)

                # Finally, I can add all of these individuals to the "UnitUlcers" section.
                month.getElementsByTagName("UnitUlcers")[0].appendChild(HAPI_patient_node)

            # I am going to do the same for each listed Restraint patient
            for i in range(len(Restraint_patients)):
                # and create a "RestraintPatient" node for each individual patient
                Restraint_patient_node = tree.createElement("RestraintPatient")
                for data in individual_restraint:
                    # I create an element for each individual's metrics
                    data_name = tree.createElement(data)

                    if data in ["xGenderFID", "xRestraintInUseFID", "xRestraintCategoryFID"] and pd.isna(Restraint_patients[data].iloc[i]):
                        data_name.appendChild(tree.createTextNode(''))
                    elif data in ["PatientAge", "xRestraintUlcerAgeMonthsFID",
                                  "xRestraintUlcerAgeDaysFID"] and Restraint_patients[data].iloc[i] == 0:
                        data_name.setAttribute('xsi:nil', 'true')
                    elif pd.isna(Restraint_patients[data].iloc[i]):
                        data_name.setAttribute('xsi:nil', 'true')
                    else:
                        # and I add each one to that individual's "UnitUlcer" node
                        data_name.appendChild(tree.createTextNode(str(Restraint_patients[data].iloc[i])))

                    Restraint_patient_node.appendChild(data_name)

                # Finally, I can add all of these individuals to the "UnitUlcers" section.
                month.getElementsByTagName("RestraintPatients")[0].appendChild(Restraint_patient_node)            

    # Now, we need to write and save this XML file as its own NDNQI Quarter and Year and Hospital to upload
    # into NDNQI.
    tree.writexml(open('Q{}_{} {} HAPI_Restraint Survey.xml'.format(Calendar_Quarter, Calendar_Year,Selected_Hosp), 'w'))
```

```python
# The variables needed for this function are the: Hospital abbreviation, Hospital Temp XML file,
# Calendar year being submitted, and Calendar quarter being submitted
def EUH_Indiv_Restraint_XML(Selected_Hosp,Temp_XML,Calendar_Year,Calendar_Quarter):
    # This section is to insert the individual HAPI and Restraint patients
    
    # This Hospital's last Quarter's individual HAPI details
    Hosp_individual_restraints = EUH_Restraint_Pts[EUH_Restraint_Pts.Hospital == Selected_Hosp]

    individual_restraint = ["RestraintIdentifier", "PatientAge", "xRestraintUlcerAgeMonthsFID", 
    "xRestraintUlcerAgeDaysFID","xGenderFID", "xRestraintInUseFID", "TypeLimb", "TypeVest", "TypeOther",
    "xRestraintCategoryFID", "JustifyFall", "JustifyRemoveEquipment", "JustifyHarmSelf", "JustifyHarmOthers", 
    "JustifyOther", "JustifyUnknown"]


    # This is the temporary xml file that has already had the Unit Census and date section filled
    tree = xml.dom.minidom.parse(Temp_XML)
    unit_nodes = tree.getElementsByTagName("UnitYrQtr")

    # This loops through each listed unit in our xml file
    for idx, unit in enumerate(unit_nodes):
        each_unit = unit.getElementsByTagName('UnitName')[0].childNodes[0].nodeValue

        # Then we loop through each listed month
        month_nodes = unit.getElementsByTagName("UnitMonth")
        for idx2, month in enumerate(month_nodes):

            # Then I grab each patient from Hosp_individual_HAPI data for this unit
            Restraint_patients = Hosp_individual_restraints[Hosp_individual_restraints.UnitName == each_unit]

            # I am going to do the same for each listed Restraint patient
            for i in range(len(Restraint_patients)):
                # and create a "RestraintPatient" node for each individual patient
                Restraint_patient_node = tree.createElement("RestraintPatient")
                for data in individual_restraint:
                    # I create an element for each individual's metrics
                    data_name = tree.createElement(data)

                    if data in ["xGenderFID", "xRestraintInUseFID", "xRestraintCategoryFID"] and pd.isna(Restraint_patients[data].iloc[i]):
                        data_name.appendChild(tree.createTextNode(''))
                    elif data in ["PatientAge", "xRestraintUlcerAgeMonthsFID",
                                  "xRestraintUlcerAgeDaysFID"] and Restraint_patients[data].iloc[i] == 0:
                        data_name.setAttribute('xsi:nil', 'true')
                    elif pd.isna(Restraint_patients[data].iloc[i]):
                        data_name.setAttribute('xsi:nil', 'true')
                    else:
                        # and I add each one to that individual's "UnitUlcer" node
                        data_name.appendChild(tree.createTextNode(str(Restraint_patients[data].iloc[i])))

                    Restraint_patient_node.appendChild(data_name)

                # Finally, I can add all of these individuals to the "UnitUlcers" section.
                month.getElementsByTagName("RestraintPatients")[0].appendChild(Restraint_patient_node)            

    # Now, we need to write and save this XML file as its own NDNQI Quarter and Year and Hospital to upload
    # into NDNQI.
    tree.writexml(open('Q{}_{} {} Restraint_Only Survey.xml'.format(Calendar_Quarter, Calendar_Year,Selected_Hosp), 'w'))
```
